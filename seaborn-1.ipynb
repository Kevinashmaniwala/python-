```python
import numpy as np
import seaborn as sns

sns.set(style="white")

# Generate a random univariate dataset
rs = np.random.RandomState(10)
d = rs.normal(size=100)

# Define the colors to use
colors = ["r", "g", "b"]

# Plot a histogram with multiple colors
sns.distplot(d, kde=True, hist=True, bins=10,
             rug=True,hist_kws={"alpha": 0.3,
                                "color": colors[0]},
             kde_kws={"color": colors[1], "lw": 2},
             rug_kws={"color": colors[2]})
```

```python
# Python program to illustrate 
# Plotting categorical scatter 
# plots with Seaborn 

# importing the required module 
import matplotlib.pyplot as plt 
import seaborn as sns 

# x axis values 
x =['sun', 'mon', 'fri', 'sat', 'tue', 'wed', 'thu'] 

# y axis values 
y =[2, 3.5, 5, 7, 4.5, 6.7, 8.9] 

# plotting strip plot with seaborn 
ax = sns.stripplot(x, y); 

# giving labels to x-axis and y-axis 
ax.set(xlabel ='Days', ylabel ='Amount_spend') 

# giving title to the plot 
plt.title('My first graph'); 

# function to show plot 
plt.show() 
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Create a custom array for total bills
# For example, let's create an array of total bills
total_bills = np.random.normal(loc=20, scale=5, size=100)  # Normally distributed data around $20

# Create a DataFrame from the custom array
# For this example, we'll create a DataFrame with additional columns for demonstration
data = {
    'total_bill': total_bills,
    'smoker': np.random.choice(['Yes', 'No'], size=100),  # Randomly assigning smokers
    'time': np.random.choice(['Lunch', 'Dinner'], size=100)  # Randomly assigning time
}

# Convert the dictionary to a DataFrame
df = sns.load_dataset('tips')  # Load the tips dataset to get the structure
df = df.iloc[:0]  # Clear the DataFrame structure
df = df.assign(**data)  # Assign the custom data

############# Main Section ############# 
# Form a facetgrid using rows for smokers and columns for time
graph = sns.FacetGrid(df, row='smoker', col='time') 
# Map the above form facetgrid with some attributes
graph.map(plt.hist, 'total_bill', bins=15, color='orange') 
# Show the object
plt.show() 
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# User-defined arrays
total_bills = np.array([15.50, 22.00, 18.75, 30.00, 25.50, 10.00, 40.00, 35.25, 20.00, 12.50,
                         28.00, 33.50, 19.75, 24.00, 15.00, 27.50, 22.50, 18.00, 29.00, 21.00])
smoker_status = np.array(['Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'Yes', 'No',
                          'Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'Yes', 'No'])
meal_time = np.array(['Lunch', 'Lunch', 'Dinner', 'Dinner', 'Lunch', 'Dinner', 'Lunch', 'Dinner',
                      'Lunch', 'Dinner', 'Lunch', 'Dinner', 'Lunch', 'Dinner', 'Lunch', 'Dinner',
                      'Lunch', 'Dinner', 'Lunch', 'Dinner'])

# Create a DataFrame from the user-defined arrays
data = {
    'total_bill': total_bills,
    'smoker': smoker_status,
    'time': meal_time
}

df = pd.DataFrame(data)

############# Main Section ############# 
# Form a facetgrid using rows for smokers and columns for time
# Set height and aspect to control the size of the graphs
graph = sns.FacetGrid(df, row='smoker', col='time', height=4, aspect=1) 
# Map the above form facetgrid with some attributes
graph.map(plt.hist, 'total_bill', bins=5, color='orange') 
# Show the object
plt.show() 
```
